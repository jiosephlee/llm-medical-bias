```python
import pandas as pd 

counterfactual_df = pd.read_csv('triage_counterfactual.csv')
patients = pd.read_csv('hosp_patients_public.csv')

# I want to merge the two dataframes on the subject_id column and add anchor_age to the counterfactual df
# Merge counterfactual_df with patients df to add anchor_age
counterfactual_df = counterfactual_df.merge(
    patients[['subject_id', 'anchor_age']], 
    on='subject_id',
    how='left'
)
edstays = pd.read_csv('edstays.csv')
# Merge counterfactual_df with edstays df to add arrival_transport
counterfactual_df = counterfactual_df.merge(
    edstays[['stay_id', 'arrival_transport']], 
    on='stay_id',
    how='left'
)
```
